```python
import chardet
import pandas as pd


file = 'C:/Users/emielou/Desktop/Recommendation/Man_Dataset.csv'  


def loadcsv(file):
    
    df = pd.DataFrame()
    
    if file.endswith(".csv"):
            with open(file, "rb") as fp:
                encoding = chardet.detect(fp.read())["encoding"]
            df = pd.concat(
                [
                    df,
                    pd.read_csv(file, encoding=encoding),
                ],
                axis=0,
            )
    return df

df = loadcsv(file)
```
